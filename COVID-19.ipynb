```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.worldometers.info/coronavirus/')
```

```python
response.status_code
```

```python
soup1 = BeautifulSoup(response.text, 'html.parser')#'lxml')  # build tree
```

```python
s= soup1.find_all('tr')
str1 = ""      
# traverse in the string   
for ele in s:
    ss=str(ele)
    str1 += ss
    
soup3 = BeautifulSoup(str1, 'html.parser')
```

```python
len(soup3.find_all('td', attrs={'class':""}))
```

```python
item_list= []
for mp in range(3312):
    item_list.append(soup3.find_all('td', attrs={'class':""})[mp].string)

len(item_list)
```

```python
countries = []
e=0
for i in range(3312):
    if e < 3312:
        countries.append(item_list[e])
        e+=9
```

```python
len (countries)
```

```python
total_cases = []
e=1
for i in range(3312):
    if e < 3313:
        total_cases.append(item_list[e])
        e+=9
```

```python
new_cases = []
e=2
for i in range(3312):
    if e < 3313:
        new_cases.append(item_list[e])
        e+=9
```
